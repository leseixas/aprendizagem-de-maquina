```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
xs = [100*random.random() for i in range(20)]
xs.sort()
print(xs)
```

```python
def f(x: float):
    if x<=20:
        return np.abs(random.random()*5+0.5*x)
    elif (x>20 and x<=40):
        return np.abs(random.random()*4+40-0.4*x)
    elif (x>40 and x<=75):
        return 100-random.random()*2
    elif (x>75):
        return np.abs(random.random()*20-0.05*x)
    
```

```python
ys = [f(x) for x in xs]
```

```python
plt.scatter(xs, ys)
plt.show()
```

```python
df = pd.DataFrame(xs, columns=['x'])
```

```python
df['y'] = ys
```

```python
df = pd.read_csv("../exemplo_03.csv")
```

```python
X, y = df[['x']], df[['y']]
```

```python
# df.to_csv("exemplo_03.csv", index=False)
```

```python
import seaborn as sns
sns.set_style("ticks")

plt.scatter(X,y, s=100, color="#2c7fb8", edgecolors="black", linewidths=1.5, alpha=0.7)

plt.ylim(-5,105)
plt.xlim(0,100)
plt.xlabel('$x$', fontsize=18)  
plt.ylabel('$y$', fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=18, width=2)
plt.gca().set_aspect('equal', adjustable='box')

for spine in plt.gca().spines.values():
    spine.set_linewidth(2)

plt.savefig("../../graficos/exemplo03/valores.png", dpi=600, format='png', bbox_inches='tight')
plt.show()
```
