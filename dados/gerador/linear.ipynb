```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import random
```

```python
xs = np.array([75*np.random.random() for i in range(15)])
xs.sort()
print(xs)
```

```python
noise= np.array([10*np.random.random() for _ in range(15)])
```

```python
ys = 0.8*xs+noise
```

```python
# pandas
df = pd.DataFrame(xs, columns=['x'])
df['y'] = ys
# print(df)

df.to_csv("../exemplo_linear.csv") # saving DataFrame

# Loading DataFrame
# df = pd.read_csv("../exemplo_linear.csv")
```

```python
import seaborn as sns
sns.set_style("ticks")

plt.scatter(xs,ys, s=100, color="#2c7fb8", edgecolors="black", linewidths=1.5, alpha=0.8)

plt.ylim(0,100)
plt.xlim(0,100)
plt.xlabel('$x$', fontsize=18)  
plt.ylabel('$y$', fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=18, width=2)
plt.gca().set_aspect('equal', adjustable='box')

for spine in plt.gca().spines.values():
    spine.set_linewidth(2)

plt.savefig("../../graficos/exemplo_linear/valores.png", dpi=600, format='png', bbox_inches='tight')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X, y = df[['x']], df[['y']]

model = LinearRegression()

model.fit(X, y)

```

```python
a, b = model.coef_, model.intercept_
print(a,b)
```

```python
linear_model = a*np.arange(0,101,5) + b

print(linear_model)
```

```python
sns.set_style("ticks")

plt.scatter(xs,ys, s=100, color="#2c7fb8", edgecolors="black", linewidths=1.5, alpha=0.8)
plt.plot(np.arange(0,101,5), linear_model.ravel(), color='black', linestyle='--', linewidth=2.5)


plt.ylim(0,100)
plt.xlim(0,100)
plt.xlabel('$x$', fontsize=18)  
plt.ylabel('$y$', fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=18, width=2)
plt.gca().set_aspect('equal', adjustable='box')

for spine in plt.gca().spines.values():
    spine.set_linewidth(2)

plt.savefig("../../graficos/exemplo_linear/linear_regression.png", dpi=600, format='png', bbox_inches='tight')
plt.show()
```
