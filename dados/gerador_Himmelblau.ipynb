```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
def my_random():
    return 10*random.random() - 5
```

```python
s = np.array( [ [my_random(), my_random()] for i in range(200)] )
```

# [Himmelblau's function](https://en.wikipedia.org/wiki/Himmelblau%27s_function)



```python
def himmelblau_log(x,y):
    z = (x**2 + y -11)**2 + (x+y**2 - 7)**2
    return np.log(z)
```

```python
h = himmelblau_log(s[:,0], s[:,1])
```

```python
df = pd.DataFrame(s, columns =['x0', 'x1'])
```

```python
df['y'] = h
```

```python
df = pd.read_csv('himmelblau_log.csv')
```

```python
print(df)
```

```python
import seaborn as sns
sns.set_theme(style='ticks', context='talk')
plt.scatter(df['x0'], df['x1'], c=df['y'], s=80, edgecolors="black", cmap=plt.cm.RdYlBu)
plt.xlim(-5.5,5.5)
plt.ylim(-5.5,5.5)
plt.xlabel('$x_0$', fontsize=18)  
plt.ylabel('$x_1$', fontsize=18) 
plt.tick_params(axis='both', which='major', labelsize=18)
cbar = plt.colorbar()
cbar.set_label('y')
plt.gca().set_aspect(1)
plt.savefig('datasets_dt_himmelblau_01.png', dpi=600, format='png', bbox_inches='tight', transparent=True)
plt.show()
```

```python
df.to_csv('himmelblau_log.csv', index=False)
```
